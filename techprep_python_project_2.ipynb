```python
# CSV paths for 
csv_path_input = 'Data/Grades_Short.csv'
csv_path_output = 'Data/Grades_Short_Modified.csv'


# Define a function that takes a column of numbers (final grade)
# and assigns a matching letter grade. 
# Function arguments:
# final_grade_column = a column of numbers we match to letter grades
# df = the dataframe that we are going to add the new column to
def calculate_letter_grade(final_grade_column,df):
    data = []
    for final_grade in final_grade_column:
        match final_grade:
            case final_grade if final_grade >= 90:
                data.append("A+")
            case final_grade if final_grade >= 80 and final_grade < 90:
                data.append("A")
            case final_grade if final_grade >= 70 and final_grade < 80:
                data.append("B")
            case final_grade if final_grade >= 60 and final_grade < 70:
                data.append("C")
            case final_grade if final_grade > 55 and final_grade < 60:
                data.append("D")
            case final_grade if final_grade < 55:
                data.append("F")
        
    df['Letter_Grade'] = data
    return df

# ------------------------ #
# Main part of the program.#
# ------------------------ #

# Prepare the workspace
import pandas as pd
df = pd.read_csv(csv_path_input)

# Select the grade columns and calculate the average (mean) of each row
df['Final_Grade'] = df[['Assignment_1','Assignment_2','Quiz_1','Quiz_2','Mid_Term_Exam','Final_Exam']].mean(axis=1)

# Use the function we made earlier to calculate the letter grade and
# add it as a new column
calculate_letter_grade(df['Final_Grade'],df)
print(df)
df.to_csv(csv_path_output)
```
